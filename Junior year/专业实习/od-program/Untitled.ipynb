```python
import pandas as pd
```

```python
data1 = pd.read_excel('ans.xlsx',sheet_name=0)
data1.head()
```

```python
data2 = pd.read_excel('ans.xlsx',sheet_name=1)
data2.head()
```

```python
possion=data2.iloc[1,:]
```

```python
possion
```

```python
# data1.iloc[:,1:19].
data1.index = data1['OD']
```

```python
ans = data1.iloc[:,1:19]
```

```python
ans
```

```python
Y = ans['上车人数']
Y
```

```python
for j in range(2,19):
    syp = sum(Y[k]*possion[j-k] for k in range(1,j))
    for i in range(1,j):
        ans.loc[i,j] = int((ans.loc['下车人数',j]*Y[i]*possion[j-i])/syp)
        Y[i] = Y[i]-ans.loc[i,j]
```
