```python
import pandas as pd
```

```python
table = pd.read_excel('demand.xlsx')
table.head()
```

```python
Q = 10
N = 30
table['n00'] = 9999
table.iloc[0,2:7]=0
table.head()
```

```python
for i in range(len(table)):
    # 当0s时已初始化完毕
    if i==0:
        continue
        
    for j in range(2,6,1):
        q1 = min(10,table.iloc[i-1,j-1],2/3*(N-table.iloc[i-1,j])) #q1为流入，q2为流出
        if i<=9 and j==5: #此为红灯状态
            q2 = 0
        elif j==5: #红灯结束但为最后一个cell
            q2 = min(10,table.iloc[i-1,j])
        else: #红灯结束其它cell
            q2 = min(10,table.iloc[i-1,j],2/3*(N-table.iloc[i-1,j+1]) )
        table.iloc[i,j] = table.iloc[i-1,j]+q1-q2
    # 统计流出量
    table.iloc[i,6] = table.iloc[i-1,6]+q2

    
```
