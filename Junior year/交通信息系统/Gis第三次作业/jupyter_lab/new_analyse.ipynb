```python
import pandas as pd
```

```python
data = pd.read_excel('../郫都数据/edges.xlsx')
data.head()
```

```python
w_data = pd.read_excel('../郫都数据/w_data.xlsx')
w_data.head()
```

```python
w_dict={}
for i,j in w_data.iterrows():
    w_dict[j[0]]=j[1]
w_dict
```

```python
w_dict['tertiary']
```

```python
for i,r in data.iterrows():
    s = r[7]
    data.loc[i,'w']=w_dict[s]
```

```python
data.head()
```

```python
new = pd.DataFrame()
new['u'] = data['u']
new['v'] = data['v']
new['w'] = data['w']
new.head()
```

```python
nodes = {}
for i, r in new.iterrows():
    nodes[int(r[1])] = 0
    nodes[int(r[0])] = 0
```

```python
for i, r in new.iterrows():
    nodes[int(r[1])] = nodes[int(r[1])] + r['w']
    nodes[int(r[0])] = nodes[int(r[0])] + r['w']
```

```python
nodes
```

```python
output = pd.DataFrame([nodes]).T
output = output.reset_index().rename(columns={'index':'id',0:'w'})
```

```python
output.head()
```

```python
output.to_csv('output/node_w.csv')
```

```python
struct_network = pd.DataFrame()
struct_network['highway'] = data['highway']
struct_network['length'] = data['length']
struct_network.head()
```

```python
avg_length = struct_network['length'].mean()
avg_length
```

```python
str_lenth = {}
str_lenth['primary']=0
str_lenth['secondary']=0
str_lenth['tertiary']=0
str_lenth['trunk']=0
```

```python
for i,w in struct_network.iterrows():
    if w[0] =='primary' or w[0] =='secondary' or w[0] =='tertiary' or w[0] =='trunk':
        str_lenth[w[0]] = str_lenth[w[0]]+w[1]
str_lenth
```

```python
str_lenth = pd.DataFrame([str_lenth]).T
str_lenth = str_lenth.reset_index().rename(columns={'index':'Rank',0:'Lenth_sum'})
str_lenth
```

```python
L_sum = str_lenth['Lenth_sum'].sum()
str_lenth['Proportion'] = str_lenth['Lenth_sum']/L_sum
str_lenth
```
