```python
import pandas as pd
import numpy as np
from geopy.distance import geodesic
import datetime
import random
```

```python
def get_v(data_id):
    if len(data_id)<2:
        print(data_id['id'][0],"数据残缺")
        return data_id
    data_id['distance'][0] = 0
    data_id['t'][0] = 0
    data_id['v'][0] = 0
    for i in range(1,len(data_id)):
        lat1 = data_id['latitude'][i-1]
        lan1 = data_id['longitude'][i-1]
        lat2 = data_id['latitude'][i]
        lan2 = data_id['longitude'][i]
        distance = geodesic((lat1, lan1), (lat2, lan2)).m
        data_id['distance'][i] = distance
        
        str_p1 = data_id['time'][i-1]
        dateTime_p1 = datetime.datetime.strptime(str_p1,'%Y/%m/%d %H:%M:%S')
        str_p2 = data_id['time'][i]
        dateTime_p2 = datetime.datetime.strptime(str_p2,'%Y/%m/%d %H:%M:%S')
        dh = (dateTime_p2-dateTime_p1).seconds/3600
        data_id['t'][i] = dh
        if dh!=0:
            v = distance/dh
            data_id['v'][i] = v
        else:
            print('error2')
    return data_id
```
